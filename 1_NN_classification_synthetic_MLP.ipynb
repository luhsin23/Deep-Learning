### Define our Pytorch Datasets and Dataloaders
You will need to make some Datasets and Dataloaders for the Sprials data and the XOR data. Datasets are a particular class used in Pytorch.  Here is an example of how to make a custom Dataset from one of the CS449 github repo functions. We wrap Datasets in Dataloaders. The Dataloader is iterable and is used to dole out batches of training data to our neural network models. For more on datasets and dataloaders, see [this pytorch tutorial](https://pytorch.org/tutorials/beginner/basics/data_tutorial.html)

````python
from numpy import pi


class TwoGaussiansDataset(torch.utils.data.Dataset):
    """TwoGaussians"""

    def __init__(self, examples_per_class, distance_between_means=3):
        """create a 2-dimensional set of points, where half the points are drawn from
        one Gaussian and the other half are drawn from a different Gaussian

        PARAMETERS
        ----------
        examples_per_class      An integer determining how much data we'll generate

        distance_between_means  Distance between the means of the two Gaussians.
        """

        data, label = make_two_gaussians_data(examples_per_class, distance_between_means)

        self.data =  torch.tensor(data).to(dtype=torch.float32)
        self.label = torch.tensor(label).to(dtype=torch.float32)
    def __len__(self):
        return len(self.label)

    def __getitem__(self, idx):
        return self.data[idx], self.label[idx]

# Make the spirals data
loader = torch.utils.data.DataLoader(TwoGaussiansDataset(examples_per_class=200), batch_size=400, shuffle=True)

# plot it
data,labels = next(iter(loader))
plot_data(data,labels)

print(data.shape)
````

### Helper functions for training
This is where you would define functions for a training loop, a testing loop, and for runing before/after trials.

````python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(2,24)
        self.fc2 = nn.Linear(24,16)
        self.fc3 = nn.Linear(16,1)

    def forward(self, x):
        x = torch.tanh(self.fc1(x))
        x = torch.tanh(self.fc2(x))
        x = torch.tanh(self.fc3(x))
        return x
````

````python
def train_model(model, optimizer, data, target):
    model.train()
    optimizer.zero_grad()
    output = model(data).squeeze()
    loss = torch.mean(torch.clamp(1 - output * target, min=0))
    loss.backward()
    torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
    optimizer.step()
    return loss.item()
def test_model(model, data, target):
    model.eval()
    with torch.no_grad():
        output = model(data).squeeze()
        output = torch.clamp(output, min=-10, max=10) # Clamp output to avoid explosions
        loss = torch.mean(torch.clamp(1 - output * target, min=0))
        prediction = torch.sign(output)
        prediction[prediction == 0] = 1
        accuracy = torch.mean((prediction == target).float())
    return accuracy.item(), loss.item()
````

````python
import numpy as np
import torch

def try_pretraining(pretrain_data_loader, downstream_data_loader, num_trials=100, max_epoch=500, threshold=0.95, verbose=True):
    '''
    Performs pretraining comparison with num_trials trials and records the crossing epochs for both pretrained and regular models.

    INPUTS
    ----------------------------------------------------------------
    pretrain_data_loader    : a data loader for pretraining data
    downstream_data_loader  : a data loader for downstream data
    num_trials              : number of trials to run
    max_epoch               : maximum number of epochs to train for
    threshold               : accuracy threshold to cross to terminate the training
    verbose                 : If true, more information is printed

    RETURNS
    ----------------------------------------------------------------
    modelPretrain           : the pretrained model from the last trial
    modelRegular            : the regular (not pretrained) model from the last trial
    pretrain_crossing_epoch : array of crossing epochs for pretrained models
    regular_crossing_epoch  : array of crossing epochs for non-pretrained models
    '''

    pretrain_crossing_epoch = np.ones(num_trials) * -1
    regular_crossing_epoch = np.ones(num_trials) * -1

    for trial in range(num_trials):
        if verbose:
            print(f"Starting trial {trial+1}/{num_trials}")

        # Initialize models
        modelPretrain = Net()
        modelRegular = Net()
        
        # Initialize weights
        def initialize_weights(model):
            for m in model.modules():
                if isinstance(m, torch.nn.Linear):
                    torch.nn.init.xavier_uniform_(m.weight)  # Xavier initialization for weights
                    torch.nn.init.zeros_(m.bias)  
                    
        initialize_weights(modelPretrain)
        initialize_weights(modelRegular)

        # Copy weights from modelRegular to modelPretrain
        modelPretrain.load_state_dict(modelRegular.state_dict())

        # Define optimizers
        optimizerPretrain = torch.optim.SGD(modelPretrain.parameters(), lr=0.01, momentum=0.9)
        optimizerRegular = torch.optim.SGD(modelRegular.parameters(), lr=0.01, momentum=0.9)

        # Step 1: Pretrain the pretrained model
        for epoch in range(max_epoch):
            for data, target in pretrain_data_loader:
                train_model(modelPretrain, optimizerPretrain, data, target)

        # Step 2: Train both models on the downstream task
        for epoch in range(max_epoch):
            for data, target in downstream_data_loader:
                # Train both models
                train_model(modelPretrain, optimizerPretrain, data, target)
                train_model(modelRegular, optimizerRegular, data, target)

            # Validate both models
            validation_data, validation_target = next(iter(downstream_data_loader))
            pretrain_accuracy, _ = test_model(modelPretrain, validation_data, validation_target)
            regular_accuracy, _ = test_model(modelRegular, validation_data, validation_target)

            # Record crossing epochs
            if pretrain_crossing_epoch[trial] == -1 and pretrain_accuracy > threshold:
                pretrain_crossing_epoch[trial] = epoch
                if verbose:
                    print(f"Pretrained model crossed threshold in trial {trial+1} at epoch {epoch}.")

            if regular_crossing_epoch[trial] == -1 and regular_accuracy > threshold:
                regular_crossing_epoch[trial] = epoch
                if verbose:
                    print(f"Regular model crossed threshold in trial {trial+1} at epoch {epoch}.")

            # If both models have crossed the threshold, stop early
            if pretrain_crossing_epoch[trial] > -1 and regular_crossing_epoch[trial] > -1:
                break

    return modelPretrain, modelRegular, pretrain_crossing_epoch, regular_crossing_epoch

````

````python
# Spiral to XOR Pretraining
modelPretrain_spiral, modelRegular_xor, pretrain_crossing_epoch_spiral, regular_crossing_epoch_xor = try_pretraining(
    pretrain_data_loader=train_loader,  # Pretrain on Spiral
    downstream_data_loader=train_loader_XOR,  # Downstream task: XOR
    num_trials=20,
    max_epoch=500,
    threshold=0.95,
    verbose=True
)

# Analyze results
print("Spiral Pretraining -> XOR:")
print("Pretrained model crossing epochs:", pretrain_crossing_epoch_spiral)
print("Regular model crossing epochs:", regular_crossing_epoch_xor)
````
